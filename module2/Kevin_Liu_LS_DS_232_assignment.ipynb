```python
# Add an OLS (Ordinary Least Squares) trendline,
# to see how the outliers influence the "line of best fit"
px.scatter(tridf, x='GROSS_SQUARE_FEET', y ='SALE_PRICE', trendline='ols')
```

```python
# Look at sales for more than $35 million

# All are at 70 Vestry Street
# All but one have the same SALE_PRICE & SALE_DATE
# Was the SALE_PRICE for each? Or in total?
# Is this dirty data?
```

```python
# At first glance it looks like it could be dirty data, as they all have the
# same price although they have varying gross square footage. although with the
# question if this is for each or in total I think it's infinitely more likely
# for it to be in total as they're all on the same street and  built in the same
# year
tridf.loc[tridf['SALE_PRICE'] > 35000000]
```

```python
# Make a judgment call:
# Keep rows where sale price was < $35 million

# Check how many rows you have now. (Should be down to 90 rows.)
tridf = tridf.loc[tridf['SALE_PRICE'] < 35000000]
print(tridf.shape)
tridf
```

```python
# Now that you've removed outliers,
# Look again at a scatter plot with OLS (Ordinary Least Squares) trendline
px.scatter(tridf, x='GROSS_SQUARE_FEET', y ='SALE_PRICE', trendline='ols')
```
